```python
consumer_key = "---------------------"
consumer_secret_key = "-----------------------------------------------"
access_token = "---------------------------------------------------"
access_secret = "---------------------------------"

auth = OAuthHandler(consumer_key, consumer_secret_key )
auth.set_access_token(access_token ,access_secret )

api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
d = {}
result = []
for tweet in tweepy.Cursor(api.search, q = '#FIFAworldcup2018', lang = "en" ).items(200):
    result.append(tweet.text)
    
```

```python
import re
result2 = []
for i in range(0, len(result)):
    txt = ' '.join(word for word in result[i].split() if not word.startswith("https:"))  #removes link
#     rmv = re.findall(r'[a-zA-Z]+', txt)                           #removes other lang text and digits
    result2.append(txt)
    result2 = [y for y in result2 if y != '']                     #removes null value

    
d['Tweet'] = result2
```
